```python
def getDataset():
    
    # Function getBoundaries
    def getBoundaries(fileName: str):
        
        with open(os.path.join("dataset", "annotations", fileName), 'r') as f:
            data = f.read()

        bs = bs4.BeautifulSoup(data)

        xmin = int(bs.find("xmin").get_text())
        ymin = int(bs.find("ymin").get_text())
        xmax=  int(bs.find("xmax").get_text())
        ymax = int(bs.find("ymax").get_text())
        
        return xmin, ymin, xmax, ymax

    
    df = pd.DataFrame(columns=["Name", "img", "xmin", "ymin", "xmax", "ymax"])
    for imgName in os.listdir("dataset/images"):
        image = cv2.imread(os.path.join("dataset", "images", imgName))
        image = cv2.cvtColor( image , cv2.COLOR_BGR2RGB )

        name = imgName.split(".")[0]
        xmlFileName = name+".xml"
        
        xmin, ymin, xmax, ymax = getBoundaries(xmlFileName)
        
        df = df.append( {"Name":name ,"img":image, "xmin":xmin, "ymin":ymin ,"xmax":xmax, "ymax":ymax }, ignore_index=True )
        
        
    return df
     
dataDF = getDataset()
    
    
```
